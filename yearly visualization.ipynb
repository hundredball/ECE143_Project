```python
import pandas as pd
import numpy as np
import geopandas as gpd
import geoplot as gpt
import matplotlib.pyplot as plt
```

```python
with open('A:/UCSD/Winter 2020/ECE 143/project/data/SanDiego_zipcode.txt', 'r') as f:
    line = f.readline()
    zips = line.strip().replace(',', '').split()
county = gpd.read_file('san_diego_zip_codes.geojson')
san_diego = county[county['zip'].isin(zips)]
san_diego = san_diego.reset_index()
san_diego.rename(columns = {'zip':'address_zip'}, inplace = True)
```

```python
_07 = pd.read_csv('A:/UCSD/Winter 2020/ECE 143/project/data/fd_incidents_2007_datasd_v1.csv')
_07['address_zip'] = _07['address_zip'].fillna(0.0).astype(int).astype(str)
_07_incidents = pd.pivot_table(_07, values='incident_number', index=['address_zip'], columns=[], aggfunc=np.ma.count, fill_value=0)
_07_san_diego = san_diego.merge(_07_incidents, on = 'address_zip', how = 'inner')
_07_san_diego['incident_number'] = _07_san_diego['incident_number'].fillna(0.0).astype(int)
_07_san_diego.plot(column = 'incident_number', scheme = 'quantiles', legend=True, cmap = 'OrRd', figsize = (6, 12))
plt.axis('off')
plt.savefig('A:/UCSD/Winter 2020/ECE 143/project/07 colormap.jpg')
```
