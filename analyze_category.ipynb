```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import dataloader
```

```python
years = range(2007, 2020)
fire_df = dataloader.load(years)
```

```python
category = fire_df['call_category'].value_counts()
fig, ax = plt.subplots(figsize=(30,12))
#fig.figsize = (30,12)
#fig.figure(figsize=(30,12))
ax.bar(category.index, category.values)
ax.yaxis.set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.title('Call category over %d-%d'%(years[0],years[-1]))
plt.savefig('category.png')
```
