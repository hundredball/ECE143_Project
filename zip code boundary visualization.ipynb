```python
import geopandas as gpd
import geoplot as gpt
```

```python
ca = gpd.read_file('ca_zip_code.geojson')
print(ca)
```

```python
zips = [str(i) for i in range(91901, 92200)]
san_diego1 = ca[ca['ZCTA5CE10'].isin(zips)]
print(san_diego1)
```

```python
gpt.polyplot(san_diego1)
```
